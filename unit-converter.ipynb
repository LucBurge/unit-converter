```python
import conversions

mainSelect = 0
while mainSelect != 5:
    
    print("What would you like to convert?")
    print("\t1 - Length")
    print("\t2 - Mass")
    print("\t3 - Temperature")
    print("\t4 - Time")
    print("\t5 - Exit")
    print()
    mainSelect = int(input("Select an option(1-5): "))
    print()

    if mainSelect == 1:
        subSelect = 0
        while subSelect != 3:
            print("1 - Meters to feet")
            print("2 - Feet to meters")
            print("3 - Return to main menu")
            subSelect = int(input("Select a unit to convert or return to the main menu(1-3): "))
            print()

            if subSelect == 1:
                valueIn = int(input("Enter value in meters: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toFeet(valueIn)} feet")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")

            elif subSelect == 2:
                valueIn = int(input("Enter value in feet: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toMeters(valueIn)} meters")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")
              
            elif subSelect == 3:
                print()
            
            else:
                print("Invalid selection. Please select an option from 1-3")


    elif mainSelect == 2:
        subSelect = 0
        while subSelect != 3:
            print("1 - Kilograms to pounds")
            print("2 - Pounds to kilograms")
            print("3 - Return to main menu")
            subSelect = int(input("Select a unit to convert or return to the main menu(1-3): "))
            print()

            if subSelect == 1:
                valueIn = int(input("Enter value in kilograms: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toPounds(valueIn)} pounds")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")

            elif subSelect == 2:
                valueIn = int(input("Enter value in pounds: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toKilograms(valueIn)} kilograms")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")
              
            elif subSelect == 3:
                print()
            
            else:
                print("Invalid selection. Please select an option from 1-3")

    elif mainSelect == 3:
        subSelect = 0
        while subSelect != 3:
            print("1 - Celsius to kelvin")
            print("2 - Kelvin to celsius")
            print("3 - Return to main menu")
            subSelect = int(input("Select a unit to convert or return to the main menu(1-3): "))
            print()

            if subSelect == 1:
                valueIn = int(input("Enter value in celsius: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toKelvin(valueIn)} Kelvin")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")

            elif subSelect == 2:
                valueIn = int(input("Enter value in Kelvin: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toCelsius(valueIn)} celsius")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")
              
            elif subSelect == 3:
                print()
            
            else:
                print("Invalid selection. Please select an option from 1-3")

    elif mainSelect == 4:
        subSelect = 0
        while subSelect != 3:
            print("1 - Seconds to hours")
            print("2 - Hours to seconds")
            print("3 - Return to main menu")
            subSelect = int(input("Select a unit to convert or return to the main menu(1-3): "))
            print()

            if subSelect == 1:
                valueIn = int(input("Enter value in seconds: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toHours(valueIn)} hours")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")

            elif subSelect == 2:
                valueIn = int(input("Enter value in hours: "))
                if valueIn > 0 and valueIn < 10000:
                    print(f"This is approximately: {conversions.toSeconds(valueIn)} seconds")
                    print()
                else:
                    print("Invalid value. Please enter a number from 0-10000")
              
            elif subSelect == 3:
                print()
            
            else:
                print("Invalid selection. Please select an option from 1-3")
        
    elif mainSelect == 5:
        print("Goodbye!")
    
    else:
        print("Invalid selection. Please select an option from 1-5")
```
